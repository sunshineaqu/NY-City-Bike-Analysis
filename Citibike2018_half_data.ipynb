```python
import os
import glob
import pandas as pd
```

```python
os.chdir("Resources")
os.getcwd()
```

```python
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
```

```python
#combine all files in the list
df = pd.concat([pd.read_csv(f) for f in all_filenames ])
df.head()
```

```python
df.dtypes
```

```python
df.count()
```

```python
# drop NA
df = df.dropna(how="any")
```

```python
df.count()
```

```python
# drop the data with age < 90
df = df[df['birth year'] > 1928] 
df.count()
```

```python
# drop the data with tripduration < 60 seconds
df = df[df['tripduration'] > 60] 
df.count()
```

```python
#reset index based on sfd
df = df.sort_values("starttime", ascending=True)
df.head()
```

```python
df = df.reset_index(drop=True)
```

```python
df.head()
```

```python
# reduce the data size by dropping rows with index ending with 0 and 5 (keep 4 rows of every 5 rows, reducd ~20%)
df = df.drop(df.iloc[::2].index, 0)
```

```python
df.count()
```

```python
# create a new age column
df["age"] = 2018 - df ["birth year"] 
```

```python
# create a new trip duration column using minute unit
df["tripduration minute"] = round(df ["tripduration"]/60, 2)
```

```python
df = df.reset_index(drop=True)
df.head(10)
```

```python
df.count()
```

```python
#export to csv
df.to_csv("combined_2018_half.csv", index=False, encoding='utf-8-sig')
```
