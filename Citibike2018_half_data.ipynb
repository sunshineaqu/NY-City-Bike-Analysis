```python
import os
import glob
import pandas as pd
```

```python
# will need to download NY city bike data from https://s3.amazonaws.com/tripdata/index.html
# and save it in a subdirectory named as "Resources" first
os.chdir("Resources")
# os.getcwd()
```

```python
# use glob.glob to find all the pathnames matching the pattern .csv
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
print(all_filenames)
```

```python
#combine all files in the list
df = pd.concat([pd.read_csv(f) for f in all_filenames ])
df.head()
```

```python
df.dtypes
```

```python
df.count()
```

```python
# drop NA
df = df.dropna(how="any")
```

```python
df.count()
```

```python
# drop the data with age < 90
df = df[df['birth year'] > 1928] 
df.count()
```

```python
# drop the data with tripduration < 60 seconds
df = df[df['tripduration'] > 60] 
df.count()
```

```python
#reset index based on sfd
df = df.sort_values("starttime", ascending=True)
df.head()
```

```python
df = df.reset_index(drop=True)
```

```python
df.head()
```

```python
# reduce the data size by dropping every other row 
# iloc[START:END:STEP] to select rows starting from START+1 row to END+1 row with step=STEP
df = df.drop(df.iloc[::2].index, 0)
# df = df.iloc[::2]
```

```python
df.count()
```

```python
# create a new age column
df["age"] = 2018 - df ["birth year"] 
```

```python
# create a new trip duration column using minute unit
df["tripduration minute"] = round(df ["tripduration"]/60, 2)
```

```python
df = df.reset_index(drop=True)
df.head(10)
```

```python
df.count()
```

```python
# Ensure the subdirectory that will use to store file has been created first (os.path.exists())
# creat one if not exist (os.mkdir())
output_name = 'combined_2018_half.csv'
output_dir = '../Output'

if not os.path.exists(output_dir):
    os.mkdir(output_dir)
    
full_path = os.path.join(output_dir, output_name)    
```

```python
# export to csv
df.to_csv(full_path, index=False, encoding='utf-8-sig')
```
